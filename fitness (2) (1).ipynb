# **Project Summary -**

```python
This analysis focuses on understanding how users spend their time in different activity levels using smart device fitness data. The dataset includes information about daily steps, calories burned, sleep patterns, and activity minutes such as Sedentary, Lightly Active, Fairly Active, and Very Active minutes.

First, the data was imported and cleaned using tools like pandas in Python inside Jupyter Notebook. Data cleaning involved checking for missing values, correcting date formats, removing duplicates, and preparing the dataset for analysis.

After cleaning the data, the analysis calculated the average time participants spend in different activity categories. The results showed that participants spend a large portion of their time in sedentary activities, while the time spent in very active activities is much lower. This suggests that many users have relatively low levels of high-intensity physical activity.

To make the analysis easier to understand, a bar chart visualization was created using matplotlib. The bar chart was selected because it clearly compares multiple categories and helps easily identify which activities have the highest and lowest average minutes.

From the insights, it becomes clear that users may benefit from encouragement to increase their physical activity levels. For example, features such as activity reminders, step goals, or fitness challenges could motivate users to reduce sedentary time and increase active minutes.

For the wellness technology company Bellabeat, these insights are valuable. By understanding how users interact with their devices and how active they are during the day, Bellabeat can design better products, personalized recommendations, and targeted marketing strategies that promote healthier lifestyles.

Overall, the analysis demonstrates how smart device data can be used to understand user behavior, identify health trends, and support data-driven decisions that improve both user wellbeing and business growth.
```

# Average steps taken by steps

```python
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

avg_steps = da.groupby('weekday')['TotalSteps'].mean().reindex(order)

avg_steps
```

```python
import matplotlib.pyplot as plt

plt.figure()

avg_steps.plot(kind='bar')

plt.title('Average Steps Taken by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Average Steps')

plt.show()
```

```python
##### 1. Why did you pick the specific chart?
```

```python
A bar chart is effective when:

We want to compare categories (Monday–Sunday).

It helps identify which days have higher or lower activity.
```

```python
##### 2. What is/are the insight(s) found from the chart?
```

```python
From the analysis, we can observe:

Users tend to take more steps on weekends, especially on Saturday.

Sunday activity slightly drops compared to Saturday.

Weekday activity levels are moderate and relatively stable.

This suggests that users are more active when they have more free time.
```

```python
activity_time = {
    'Sedentary': da['SedentaryMinutes'].mean(),
    'Light Activity': da['LightlyActiveMinutes'].mean(),
    'Fair Activity': da['FairlyActiveMinutes'].mean(),
    'Very Active': da['VeryActiveMinutes'].mean()
}
```

```python
activity_df = pd.DataFrame(list(activity_time.items()), columns=['Activity Type', 'Average Minutes'])
```

```python
plt.figure()

plt.bar(activity_df['Activity Type'], activity_df['Average Minutes'])

plt.title('Average Time Spent in Different Activity Categories')
plt.xlabel('Activity Type')
plt.ylabel('Average Minutes')

plt.show()
```

```python
Why did I pick a Bar Chart?
```

```python
I chose a Bar Chart because it is one of the most effective ways to compare values across different categories. In this analysis, the categories are:

Sedentary Activity

Light Activity

Fairly Active

Very Active

A bar chart makes it easy to visually compare how much time participants spend in each activity type.

It helps quickly identify:
```
